```python
X=df[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']]
y=df[['Chance of Admit ']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.01)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

```python
d=RandomForestRegressor()
d.fit(X_train,y_train)
```

```python
y_pred=d.predict(X_test)
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```

# Accuracy

```python
from sklearn.metrics import r2_score

# Calculate the R^2 score
r2 = r2_score(y_test, y_pred)
print("R^2 Score:", r2*100)
```

##### In conclusion, I was able to achieve an accuracy of 91.74% in Prediction of Graduate Admissions from an Indian perspective using machine learning techniques in Python.
